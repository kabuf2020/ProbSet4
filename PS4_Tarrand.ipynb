```python
#use this to find the taxonKey for brown anoles

from pygbif import species
species.name_suggest(q='Anolis sagrei')

#the brown anole is 4287413
```

```python
#use this to find the taxonKey for green anoles

species.name_suggest(q='Anolis carolinensis')

#green anole is 2466939
```

```python
#grab the occurrence data for all green anoles in Florida this year

from pygbif import occurrences
newgreen = occurrences.search(taxonKey=2466939, stateProvince='Florida', year=2020)
newgreenlats= [ x['decimalLatitude'] for x in newgreen['results'] ]
newgreenlongs=[ x['decimalLongitude'] for x in newgreen['results'] ]
```

```python
#grab the occurrence data for all green anoles in Florida in 2010

oldgreen = occurrences.search(taxonKey=2466939, stateProvince='Florida', year=2010)
oldgreenlats= [ x['decimalLatitude'] for x in oldgreen['results'] ]
oldgreenlongs=[ x['decimalLongitude'] for x in oldgreen['results'] ]
```

```python
#grab the occurrence data for all brown anoles in Florida this year

newbrown = occurrences.search(taxonKey=4287413, stateProvince='Florida', year=2020)
newbrownlats= [ x['decimalLatitude'] for x in newbrown['results'] ]
newbrownlongs=[ x['decimalLongitude'] for x in newbrown['results'] ]
```

```python
#grab the occurrence data for all brown anoles in Florida in 2010

oldbrown = occurrences.search(taxonKey=4287413, stateProvince='Florida', year=2010)
oldbrownlats= [ x['decimalLatitude'] for x in oldbrown['results'] ]
oldbrownlongs=[ x['decimalLongitude'] for x in oldbrown['results'] ]
```

```python
#this will plot green and brown anoles observed in 2010
#the figure is not recognizable as Florida, but you can compare it to the 2020 figure

import matplotlib.pyplot as plt

plt.scatter(oldgreenlongs, oldgreenlats, color = "green")
plt.scatter(oldbrownlongs, oldbrownlats, color = "brown")

plt.xlim(-88, -77)
plt.ylim(24, 31)
```

```python
#this will plot green and brown anoles observed in 2020

plt.scatter(newgreenlongs, newgreenlats, color = "green")
plt.scatter(newbrownlongs, newbrownlats, color = "brown")

plt.xlim(-88, -77)
```
